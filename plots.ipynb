```python
from matplotlib.pylab import rc

import matplotlib.pyplot as plt
import numpy as np
import matplotlib

"""
logdir="logs/fit/"
import tensorboard
%load_ext tensorboard

%tensorboard --logdir logsa


"""
plot_path = "plots/"
```

```python
names = ["Naive Bayes","Logistic regression","1D-CNN","GBDT","LSTM_FCN"]
auroc = [0.861,0.889,0.899,0.980,0.910]
auroc_err = [0.02,0.02,0.03,0.01,0.02]
accuracy = [0.701,0.775,0.802,0.950,0.831]
accuracy_err = [0.03,0.04,0.03,0.008,0.04]

font = {'family': 'normal','weight': 'bold',
        'size': 25}

color_array = ['b', 'g', 'r', 'c', 'm', 'y', 'k']


matplotlib.rc('font', **font)
rc('axes', linewidth=3)




for i in range(len(auroc)):
    plt.plot(auroc[i],accuracy[i],"o",color=color_array[i],label=names[i],markersize=10)
    plt.errorbar(x=auroc[i],xerr=auroc_err[i],y=accuracy[i],yerr=accuracy_err[i],color=color_array[i],markersize=10)

plt.xlabel("AUROC")
plt.ylabel("Accuracy")

plt.legend()

fig = matplotlib.pyplot.gcf()


fig.set_size_inches(12,12)
save_path = plot_path + "labeled_AUROC_ACC_all" + ".png"

fig.savefig(save_path, dpi=150)





```
