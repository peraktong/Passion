```python
import json
# Plot Daily samples:
# labeled data:
root_path2 = "Data/Ant_labeled/"

today = []
history = []
label = []
count=0
with open(root_path2+"train_data.txt") as f:
    for line in f:
        temp = json.loads(line)
        today.append(temp["today"])
        history.append(temp["history"])
        label.append(temp["label"])
        count+=1
today = np.array(today)
history = np.array(history)
label = np.array(label).ravel()
```

```python
mask0 = label==0
mask1 = label==1
```

```python

import matplotlib
from matplotlib.pylab import rc


font = {'family': 'normal','weight': 'bold',
        'size': 25}

matplotlib.rc('font', **font)
rc('axes', linewidth=3)



plt.plot(np.nanmedian(today[mask0],axis=0),"k",label= "Today y=0")
plt.plot(np.nanmedian(today[mask1],axis=0),"r",label= "Today y=1")

plt.plot(np.nanmedian(history[mask0],axis=0),"k--",label= "History y=0")
plt.plot(np.nanmedian(history[mask1],axis=0),"r--",label= "History y=1")

fig = matplotlib.pyplot.gcf()
plt.legend()

fig.set_size_inches(18,10)

save_path = plot_path + "Labeled_mean" + ".png"

fig.savefig(save_path, dpi=200)



```
